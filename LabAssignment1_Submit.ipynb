### Q1 : Which columns (if any) are missing values?

```python
# Q1 : Which columns (if any) are missing values?
#Ans: There are no missing value in any column


# How to get the answer :

print("---------------------------------------------")
print("Question 01 : Which columns (if any) are missing values?")
print("Answer 01 :")

# Step 01 : Check for missing values in each column
missing_values = bestsellers.isnull().sum()
# missing_values       # Column ที่มี missing value จะแสดง result ด้านหลังเป็นจำนวนมากกว่า 0

# Step 02 Filter out columns that have missing values
columns_with_missing_values = missing_values[missing_values > 0]

#  Step 03 Display the result
# Use if-else to handle the output
if not columns_with_missing_values.empty:

    print("Columns with missing values:")
    for column in columns_with_missing_values.index:
        print(f"- {column}")
else:
    print("No columns with missing values.")
```

## Part 2
**Import the `bestsellers.csv` dataset and use it to answer the following questions:**

* Find the lowest User Rating in the DF
* Find the highest Price in the DF
* What is the average User Rating?
* What the average User Rating of the first 5 rows?
* What is the total (sum) of all the values in the Reviews column?





```python
bestsellers = pd.read_csv("https://raw.githubusercontent.com/TipGreenTea/Depa-24/main/bestsellers.csv")
```

### Q1 : The lowest User Rating

```python
#Find the lowest User Rating in the DF
#Ans: 3.3

# Find the minimum value in the "User Rating" column
min_user_rating = bestsellers['User Rating'].min()

# Display the result
print("The lowest User rating is ", min_user_rating)
```

- You can also view the row with lowest user rating from below code:

```python
# Retrieve the row(s) with the minimum user rating
min_rating_row = bestsellers[bestsellers['User Rating'] == min_user_rating]
min_rating_row
```

### Q2 : The highest price

```python
#Find the highest Price in the DF
#Ans: 105

# find the maximum value in "Price" column
max_price = bestsellers['Price'].max()

# Display the result
print("The highest price is ",max_price)
```

- You can also view the row with highest price from below code:

```python
# Retrieve the row(s) with the higest price
max_price_row = bestsellers[bestsellers['Price'] == max_price]
max_price_row
```

### Q3 : The average user rating

```python
#What is the average User Rating?
#Ans: 4.62


# find the average user rating
average_user_rating = bestsellers['User Rating'].mean()

# Round the result to 2 decimal places
average_user_rating_rounded = round(average_user_rating, 2)

# Display the result
print("The average user rating is ",average_user_rating_rounded)
```

### Q4 : The average user rating for first 5 rows



```python
#What the average User Rating of the first 5 rows?
#Ans: 4.7


# find the average user rating from first 5 rows
average_user_rating_first5 =  (  bestsellers['User Rating'].head()  ).mean()

# Display the result
print("The average user rating of the first 5 rows is ",average_user_rating_first5 )
```

### Q5 : The total number of reviews

```python
#What is the total (sum) of all the values in the Reviews column?
#Ans:6,574,305

sum_reviews_all = bestsellers['Reviews'].sum()
#sum_reviews_all

# Format the result with a comma as a thousands separator
formatted_sum_reviews = f"{sum_reviews_all:,}"


# Display the result
print("The total reviews are ",formatted_sum_reviews )
```


### Q4 : Create a bar plot showing the 5 authors who have the most books with a rating under 4.5


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5


# Step 1 : Filter only book with rating under 4.5
books_User_rating_less_than_4_5 = bestsellers [ bestsellers['User Rating']  < 4.5]

# Step 2 : apply data from Step 1 Using Group by function
top5_Author_most_books_UserRating_4_5 = books_User_rating_less_than_4_5 .groupby('Author')['Name'].count()
top5_Author_most_books_UserRating_4_5

# Step 3 : Sort the authors by the number of books and select the top 5
top5_authors_most_books_sorted_desc = top5_Author_most_books_UserRating_4_5 .sort_values(ascending=False).head()
top5_authors_most_books_sorted_desc

# Step 4 : plotting the bar plot using plt
plt.figure(figsize=(10, 6))
ax = top5_authors_most_books_sorted_desc.plot(kind='bar', color='skyblue')


# Add number of books at the top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')



plt.title('The 5 authors who have the most books with a rating under 4.5')
plt.xlabel('Name of Author')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
